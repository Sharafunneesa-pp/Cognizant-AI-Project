```python
#unique values and value count in categorical
exclude_columns=['transaction_id', 'timestamp', 'product_id']

for column in data.select_dtypes(include='object').columns:
    if column not in exclude_columns:
        print(f"\n{column}:", data[column].nunique())
        print(data[column].value_counts())    
```

```python
# Distribution of numerical columns
from scipy import stats
plt.figure(figsize=(10, 6))
for column in data.select_dtypes(include=['int', 'float']):
    plt.subplot(1,3,1)
    plt.title(column)
    sns.histplot(data[column], bins=20, kde=True)
    plt.subplot(1,3,2)
    plt.title(column)
    sns.boxplot(data[column])
    plt.subplot(1,3,3)
    plt.title(column)
    stats.probplot(data[column],dist="norm",plot=plt)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.figure(figsize=(20, 20))
for column in data.select_dtypes(include='object'):
    if column not in exclude_columns:
        sns.countplot(x=data[column])
        plt.title(f'Count of {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.xticks(rotation=45)
        plt.show()
```

```python
# Correlation analysis
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
#sales trend over time
# Convert timestamp column to datetime
data['timestamp'] = pd.to_datetime(data['timestamp'])

# Group by date and calculate total sales
trends = data.groupby(data['timestamp'].dt.date)['total'].sum()

# Plot sales trends over time
trends.plot(kind='line', figsize=(10, 6), xlabel='Date', ylabel='Total Sales', title='Sales Trends Over Time')
plt.show()
```

```python
# Group by category and calculate total sales
category_analysis = data.groupby('category')['total'].sum().sort_values(ascending=False)


# Print category-wise total sales
print("Category-wise Total Sales:")
print(category_analysis)
```

```python
# Group by customer_type and calculate average total sales
customer_segmentation = data.groupby('customer_type')['total'].mean().sort_values(ascending=False)


# Print average total sales for each customer segment
print("Average Total Sales by Customer Segment:")
print(customer_segmentation)
```

```python
import seaborn as sns

# Plot relationship between unit price and total sales
sns.scatterplot(x='unit_price', y='total', data=data)
plt.xlabel('Unit Price')
plt.ylabel('Total Sales')
plt.title('Price Sensitivity Analysis')
plt.show()
```

```python
payment_analysis = data.groupby('payment_type')['total'].sum().sort_values(ascending=False)


# Print total sales by payment method
print("Total Sales by Payment Method:")
print(payment_analysis)
```

```python
# Calculate average transaction value
average_transaction_value = data['total'].mean()

# Print average transaction value
print("Average Transaction Value:", average_transaction_value)
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])

# Group by date and calculate total sales
trends = data.groupby(data['timestamp'].dt.date)['total'].sum()



# Average sales per day
average_sales_per_day = trends.mean()
print(f"\nAverage Sales Per Day: {average_sales_per_day:.2f}")

# Sales growth rate 
sales_growth_rate = trends.pct_change()
average_growth_rate = sales_growth_rate.mean() * 100  
print(f"\nAverage Daily Sales Growth Rate: {average_growth_rate:.2f}%")

# Peak sales day
peak_sales_day = trends.idxmax()
peak_sales_value = trends.max()
print(f"\nPeak Sales Day: {peak_sales_day}, Total Sales: {peak_sales_value:.2f}")

# Lowest sales day
lowest_sales_day = trends.idxmin()
lowest_sales_value = trends.min()
print(f"\nPeak Sales Day: {lowest_sales_day}, Total Sales: {lowest_sales_value:.2f}")
```
