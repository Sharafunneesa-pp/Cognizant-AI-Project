**Section 2-Data Loading**

```python
import pandas as pd
```

```python
path='/content/drive/MyDrive/Colab Notebooks/sample_sales_data.csv'
df = pd.read_csv(path)
df.drop(columns=["Unnamed: 0"], inplace=True, errors='ignore')
df.head()
```

**Section 3 - Descriptive statistics**

```python
df.describe(include='all')
```

```python
df.info()
```

```python
#unique values and value count in categorical
exclude_columns=['transaction_id', 'timestamp', 'product_id']

for column in df.select_dtypes(include='object').columns:
    if column not in exclude_columns:
        print(f"\n{column}:", df[column].nunique())
        print(df[column].value_counts())
```

```python
# Group by category and calculate total sales
category_analysis = df.groupby('category')['total'].sum().sort_values(ascending=False)


# Print category-wise total sales
print("Category-wise Total Sales:")
print(category_analysis)
```

```python

category_counts = df['category'].value_counts()

# Select the top two categories
top_categories = category_counts.head(2)

print("Top Two Product Categories:")
print(top_categories)
```

```python
# Group by customer_type and calculate average total sales
customer_segmentation = df.groupby('customer_type')['total'].mean().sort_values(ascending=False)


# Print average total sales for each customer segment
print("Average Total Sales by Customer Segment:")
print(customer_segmentation)
```

```python
payment_analysis = df.groupby('payment_type')['total'].sum().sort_values(ascending=False)
print("Total Sales by Payment Method:")
print(payment_analysis)
```

```python
# Calculate average transaction value
average_transaction_value = df['total'].mean()

# Print average transaction value
print("Average Transaction Value:", average_transaction_value)
```

```python

df['timestamp'] = pd.to_datetime(df['timestamp'])

# Extract hour component from timestamp
df['hour'] = df['timestamp'].dt.hour

# Group by hour and calculate total sales or number of transactions
hourly_sales = df.groupby('hour')['total'].sum()

# Find the hour(s) with the highest total sales
peak_hours = hourly_sales.idxmax()
peak_sales = hourly_sales.max()

print("Peak Hour(s):", peak_hours)
print("Total Sales at Peak Hour(s):", peak_sales)
```

**Section 4 - Visualisation**

```python
# Distribution of numerical columns
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
plt.figure(figsize=(10, 6))
for column in df.select_dtypes(include=['int', 'float']):
    plt.subplot(1,3,1)
    plt.title(column)
    sns.histplot(df[column], bins=20, kde=True)
    plt.subplot(1,3,2)
    plt.title(column)
    sns.boxplot(df[column])
    plt.subplot(1,3,3)
    plt.title(column)
    stats.probplot(df[column],dist="norm",plot=plt)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.figure(figsize=(20, 20))
for column in df.select_dtypes(include='object'):
    if column not in exclude_columns:
        sns.countplot(x=df[column])
        plt.title(f'Count of {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.xticks(rotation=45)
        plt.show()
```

```python
# Correlation analysis
non_numeric_columns = df.select_dtypes(exclude=['float', 'int']).columns
numeric_df = df.drop(columns=non_numeric_columns)
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

df['hour'] = df['timestamp'].dt.hour

hourly_trends = df.groupby('hour')['total'].sum()

hourly_trends.plot(kind='line', figsize=(10, 6), xlabel='Hour', ylabel='Total Sales', title='Hourly Sales Trends')
plt.show()
```
